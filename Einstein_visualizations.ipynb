```python
# Import necssary libarries
import pandas as pd

# Load the two CSV files
egg_sandwiches_path = "C:\\Users\\Sridhar Reddy\\Desktop\\DATA-606\\Einstein Bros Bagels\\Egg_Sandwiches.csv"
lunch_sandwiches_path = "C:\\Users\\Sridhar Reddy\\Desktop\\DATA-606\\Einstein Bros Bagels\\Lunch_Sandwiches.csv"

# Read the CSV files
egg_sandwiches_df = pd.read_csv(egg_sandwiches_path)
lunch_sandwiches_df = pd.read_csv(lunch_sandwiches_path)

# Display the first few rows of each dataframe to understand the structure of the data
egg_sandwiches_df.head()
```

```python
lunch_sandwiches_df.head()
```

```python
# Import necssary libarries
import matplotlib.pyplot as plt

# Calculate the average price and calorie content for both datasets
egg_sandwich_avg_price = egg_sandwiches_df['Total Price($) including tax'].mean()
lunch_sandwich_avg_price = lunch_sandwiches_df['Total Price($) including tax'].mean()

egg_sandwich_avg_calories = egg_sandwiches_df['Calories'].mean()
lunch_sandwich_avg_calories = lunch_sandwiches_df['Calories'].mean()

# Create bar plots for average price and calories
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Bar plot for average price comparison
ax1.bar(['Egg Sandwiches', 'Lunch Sandwiches'], [egg_sandwich_avg_price, lunch_sandwich_avg_price], color=['skyblue', 'lightgreen'])
ax1.set_title('Average Price Comparison')
ax1.set_ylabel('Average Price ($)')

# Bar plot for average calories comparison
ax2.bar(['Egg Sandwiches', 'Lunch Sandwiches'], [egg_sandwich_avg_calories, lunch_sandwich_avg_calories], color=['orange', 'salmon'])
ax2.set_title('Average Calorie Comparison')
ax2.set_ylabel('Average Calories')

# Display the plots
plt.tight_layout()
plt.show()
```

```python
# Import necessary libraries
from collections import Counter

# Split ingredients into individual items for both datasets
egg_ingredients = [item.strip() for sublist in egg_sandwiches_df['Ingredients'].str.split(',') for item in sublist]
lunch_ingredients = [item.strip() for sublist in lunch_sandwiches_df['Ingredients'].str.split(',') for item in sublist]

# Count the frequency of ingredients
egg_ingredient_counts = Counter(egg_ingredients)
lunch_ingredient_counts = Counter(lunch_ingredients)

# Convert to dataframes for easier visualization
egg_ingredient_df = pd.DataFrame(egg_ingredient_counts.items(), columns=['Ingredient', 'Frequency']).sort_values(by='Frequency', ascending=False)
lunch_ingredient_df = pd.DataFrame(lunch_ingredient_counts.items(), columns=['Ingredient', 'Frequency']).sort_values(by='Frequency', ascending=False)

# Plot the top 10 ingredients for each dataset
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Egg sandwiches ingredient frequency
egg_ingredient_df.head(10).plot(kind='barh', x='Ingredient', y='Frequency', ax=ax1, color='skyblue')
ax1.set_title('Top 10 Ingredients in Egg Sandwiches')
ax1.set_xlabel('Frequency')

# Lunch sandwiches ingredient frequency
lunch_ingredient_df.head(10).plot(kind='barh', x='Ingredient', y='Frequency', ax=ax2, color='lightgreen')
ax2.set_title('Top 10 Ingredients in Lunch Sandwiches')
ax2.set_xlabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Plotting the top 5 highest calorie sandwiches for both categories
# Sort both datasets by calories in descending order
egg_sandwiches_sorted_by_calories = egg_sandwiches_df.sort_values(by='Calories', ascending=False)
lunch_sandwiches_sorted_by_calories = lunch_sandwiches_df.sort_values(by='Calories', ascending=False)

# Prepare data for the plot
top_egg_sandwiches = egg_sandwiches_sorted_by_calories[['Item', 'Calories']].head(5)
top_lunch_sandwiches = lunch_sandwiches_sorted_by_calories[['Item', 'Calories']].head(5)

# Plot the data
#plt.figure(figsize=(10, 6))

# Egg sandwiches bar plot
plt.barh(top_egg_sandwiches['Item'], top_egg_sandwiches['Calories'], color='skyblue', label='Egg Sandwiches')

# Lunch sandwiches bar plot
plt.barh(top_lunch_sandwiches['Item'], top_lunch_sandwiches['Calories'], color='lightgreen', label='Lunch Sandwiches')

plt.title('Top 5 Highest Calorie Sandwiches')
plt.xlabel('Calories')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Plotting the top 5 lowest calorie sandwiches for both categories

# Prepare data for the plot
lowest_egg_sandwiches = egg_sandwiches_sorted_by_calories[['Item', 'Calories']].tail(5)
lowest_lunch_sandwiches = lunch_sandwiches_sorted_by_calories[['Item', 'Calories']].tail(5)

# Plot the data
#plt.figure(figsize=(10, 6))

# Egg sandwiches bar plot
plt.barh(lowest_egg_sandwiches['Item'], lowest_egg_sandwiches['Calories'], color='skyblue', label='Egg Sandwiches')

# Lunch sandwiches bar plot
plt.barh(lowest_lunch_sandwiches['Item'], lowest_lunch_sandwiches['Calories'], color='lightgreen', label='Lunch Sandwiches')
plt.xlim(0, 1250)
plt.title('Top 5 Lowest Calorie Sandwiches')
plt.xlabel('Calories')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Import neccesary libraries
import pandas as pd
drinks_df = pd.read_csv("C:\\Users\\Sridhar Reddy\\Desktop\\DATA-606\\Einstein Bros Bagels\\Drinks.csv")
drinks_df.head()
```

```python
# Analyzing price distribution, calorie distribution, and size-based price variations

# Grouping by 'Description' to calculate average price and calories by size
avg_price_by_size = drinks_df.groupby('Description')['Total price($) including tax'].mean().reset_index()
avg_calories_by_size = drinks_df.groupby('Description')['Calories'].mean().reset_index()

# Plotting average price by drink size
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Price by size
ax1.bar(avg_price_by_size['Description'], avg_price_by_size['Total price($) including tax'], color='lightblue')
ax1.set_title('Average Price by Drink Size')
ax1.set_ylabel('Average Price ($)')

# Calories by size
ax2.bar(avg_calories_by_size['Description'], avg_calories_by_size['Calories'], color='salmon')
ax2.set_title('Average Calories by Drink Size')
ax2.set_ylabel('Average Calories')

plt.tight_layout()
plt.show()
```

```python
# Grouping by 'Item' to calculate the average calories by drink type
avg_calories_by_drink_type = drinks_df.groupby('Item')['Calories'].mean().reset_index()

# Plotting average calories by drink type
plt.figure(figsize=(12, 8))
plt.barh(avg_calories_by_drink_type['Item'], avg_calories_by_drink_type['Calories'], color='salmon')

plt.title('Average Calories by Drink Type')
plt.xlabel('Average Calories')
plt.ylabel('Drink Type')
plt.tight_layout()
plt.show()
```

```python
# Grouping by 'Item' to calculate the average price by drink type
avg_price_by_drink_type = drinks_df.groupby('Item')['Total price($) including tax'].mean().reset_index()

# Plotting average price by drink type
plt.figure(figsize=(12, 6))
plt.barh(avg_price_by_drink_type['Item'], avg_price_by_drink_type['Total price($) including tax'], color='lightblue')

plt.title('Average Price by Drink Type')
plt.xlabel('Average Price ($)')
plt.ylabel('Drink Type')
plt.tight_layout()
plt.show()
```

```python
# Split ingredients into individual items for the drinks dataset
drink_ingredients = [item.strip() for sublist in drinks_df['Ingredients'].str.split(',') for item in sublist]

# Count the frequency of each ingredient
drink_ingredient_counts = Counter(drink_ingredients)

# Convert to a dataframe for easier visualization
drink_ingredient_df = pd.DataFrame(drink_ingredient_counts.items(), columns=['Ingredient', 'Frequency']).sort_values(by='Frequency', ascending=False)

# Plot the top 10 most used ingredients
plt.figure(figsize=(10, 6))
drink_ingredient_df.head(10).plot(kind='barh', x='Ingredient', y='Frequency', color='green', legend=False)

plt.title('Top 10 Most Frequently Used Ingredients in Drinks')
plt.xlabel('Frequency')
plt.ylabel('Ingredient')
plt.tight_layout()
plt.show()
```
